```python
import matplotlib.pyplot as pl
import random
Dist=[]
First=[]
Second=[]
Third=[]
Failed=[]
grades=['Distinction','First Class','Second Class','Third Class','Failed']
for x in range(2000):
    a=random.randint(1,100)
    if(90<a<=100):
        Dist.append(a)
    elif(80<a<=90):
        First.append(a)
    elif(70<a<=80):
        Second.append(a)
    elif(60<a<=70):
        Third.append(a)
    else:
        Failed.append(a)  
a=(len(Dist),len(First),len(Second),len(Third),len(Failed))
print(a)
pl.pie(a,labels=grades,autopct='%1.1f%%')
pl.show()
```

```python
import matplotlib.pyplot as pl
import random
import pandas as pd
def CountFrequency(my_list): 
    freq = {} 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
  
    for key, value in freq.items(): 
        print ("% s : % d"%(key, value))  
        
data=pd.read_csv("D://AppleStore.csv")
b=tuple(data.track_name)
CountFrequency(b) 
```
