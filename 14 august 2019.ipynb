```python
import pandas as pd
import numpy as np
data1=pd.read_csv("D://train.csv")
```

```python
data1.head(5)
```

```python
data1.shape
```

```python
data2=data1.drop(['Name','Sex','Ticket','Cabin','Embarked'],axis=1,inplace=False)
```

```python
data2.head(5)
```

```python
data3=data2.ffill()
```

```python
data3.head(5)
```

```python
data3.isnull().sum()
```

```python
import statsmodels.formula.api as smf
logit_model = smf.logit('Survived ~ Pclass+Age+SibSp+Parch+Fare',data = data3).fit()
```

```python
logit_model.params
```

```python
logit_model.summary()
```

```python
# prediction using logit_model
y_pred1 = logit_model.predict(data3)
y_pred1
```

```python
fdata=data1.bfill()
```

```python
# Creating new column for storing predicted class of Attorney
fdata["pred_prob"] = y_pred1
fdata.head(30)

```

```python
# filling all the cells with zeroes
fdata["PValue"] = 0
fdata.head()
```

```python
# taking threshold value as 0.5 and above the prob value will be treated 
# as correct value 
fdata.loc[fdata.pred_prob >= 0.5,"PValue"] = True
fdata.head()
```

```python
# confusion matrix 
confusion_matrix = pd.crosstab(fdata['Survived'],fdata.PValue)
confusion_matrix
```

```python
# accuracy
acc = (472+152)/(472+77+190+152)
acc
np.mean(fdata.Survived == fdata.PValue)
```

```python
# taking threshold value as 0.5 and above the prob value will be treated 
# as correct value 
fdata.loc[fdata.pred_prob >= 0.6,"PValue"] = True
fdata.head()
```

```python
# confusion matrix 
confusion_matrix = pd.crosstab(fdata['Survived'],fdata.PValue)
confusion_matrix
```
